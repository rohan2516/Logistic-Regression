```python
import pandas as pd
import numpy as np
from sklearn import preprocessing as pp
```

```python
df=pd.read_csv("C:/Users/Rohan/Downloads/0000000000002429_training_titanic_x_y_train.csv")
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
def cabin(x):
    if x is np.nan:
        return 0
    else:
        return 1
```

```python
df['Cabin']=df.Cabin.apply(cabin)
```

```python
del df['Name']
del df['Ticket']
```

```python
def S(s):
    if s=='S':
        return 1
    else:
        return 0
    
def C(s):
    if s=='C':
        return 1
    else:
        return 0
    
def Q(s):
    if s=='Q':
        return 1
    else:
        return 0
```

```python
df['S']=df.Embarked.apply(S)
df['C']=df.Embarked.apply(C)
df['Q']=df.Embarked.apply(Q)
```

```python
def male(s):
    if s=='male':
        return 1
    else:
        return 0
    
def fem(s):
    if s=='female':
        return 1
    else:
        return 0
```

```python
df['male']=df.Sex.apply(male)
df['female']=df.Sex.apply(fem)
```

```python
del df['Sex']
del df['Embarked']
```

```python
df['Y']=df['Survived']
del df['Survived']
```

```python
data=df.to_numpy()
```

```python
df.describe()
```

```python
x=data[:,0:-1]
y=data[:,-1]
```

```python
sc=pp.StandardScaler()
```

```python
sc.fit(x)
```

```python
x_new=sc.transform(x)
```

```python
def sig(z):
    t=(1/(1+np.exp(-z)))
    return t
```

```python
def predict(x,theta):
    (m,n)=x.shape
    y=np.zeros((m,1))
    for i in range(m):
        p=theta[-1]
        for j in range(n):
            p+=((x[i][j])*theta[j])
        if p>=0:
            y[i]=1
        else:
            y[i]=0
    return y
```

```python
def cost(x,y,theta):
    (m,n)=x.shape
    ones=np.ones((len(x),1))
    xn=np.concatenate((x,ones),axis=1)
    xt=np.dot(xn,theta)
    h=sig(xt)
    c=(-1/m)*((y*np.log(h) + (1-y)*(np.log(1-h))).sum())
    return c
```

```python
def lr(x,y,max_iter,alpha):
    (m,n)=x.shape
    theta=np.zeros(n+1)
    ones=np.ones((len(x),1))
    xn=np.concatenate((x,ones),axis=1)
    for i in range(max_iter):
        xt=np.dot(xn,theta)
        h=sig(xt)
        theta = theta - ((alpha/m)*(np.dot(np.transpose(xn),(h-y))))
        print(cost(x,y,theta))
    return theta
```

```python
def acc(yt,yp):
    m=len(yt)
    s=0
    for i in range(m):
        if(yt[i]==yp[i]):
            s+=1
    t=(s/m)
    t=t*100
    return t
```

```python
max_iter=3000
alpha=1
theta=lr(x_new,y,max_iter,alpha)
```
